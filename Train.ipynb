```python
from ultralytics import YOLO

# Load a pretrained model.
model = YOLO('yolov8m.pt') 

# Train the model.
model.train(
    data='/content/drive/MyDrive/BRAE_428_Colab/lettuceMOT.yaml', epochs=800, 
    imgsz=[1080,810], device=0, batch=16,
    project='/content/drive/MyDrive/BRAE_428_Colab/runs', plots=True, 
    save=True
)
model.export(format='onnx')
```
